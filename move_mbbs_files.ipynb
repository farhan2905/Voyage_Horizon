# Move MBBS Location Pages
This notebook moves the newly created MBBS location HTML files into the `mbbs_abroad/` directory and verifies the move.

```python
import os
import shutil

# Source and target directories
workspace_root = os.path.abspath('.')
source_dir = os.path.join(workspace_root, 'Voyage_Horizon')
mbbs_locations_dir = source_dir
mbbs_abroad_dir = os.path.join(mbbs_locations_dir, 'mbbs_abroad')

print('Workspace root:', workspace_root)
print('Source directory:', source_dir)
print('Target mbbs_abroad directory:', mbbs_abroad_dir)
```

```python
# Ensure the mbbs_abroad directory exists
os.makedirs(mbbs_abroad_dir, exist_ok=True)
print('Created directory:', mbbs_abroad_dir)
```

```python
files_to_move = [
    'georgia.html',
    'russia.html',
    'kyrgyzstan.html',
    'kazakhstan.html',
    'uzbekistan.html',
    'egypt.html',
    'vietnam.html',
    'iraq.html',
    'nepal.html',
    'italy.html'
]

print('Files to move:')
for file in files_to_move:
    print('-', file)
```

```python
moved_files = []
failed_files = []

for filename in files_to_move:
    src = os.path.join(source_dir, filename)
    dst = os.path.join(mbbs_abroad_dir, filename)
    try:
        shutil.move(src, dst)
        moved_files.append(filename)
    except Exception as exc:
        failed_files.append((filename, str(exc)))

print('Moved files:', moved_files)
if failed_files:
    print('Failed files:')
    for filename, error in failed_files:
        print('-', filename, error)
```

```python
print('Contents of mbbs_abroad after move:')
for entry in sorted(os.listdir(mbbs_abroad_dir)):
    print('-', entry)

print('\nRoot directory check:')
for filename in files_to_move:
    print(filename, 'exists in root?' , os.path.exists(os.path.join(source_dir, filename)))
```
